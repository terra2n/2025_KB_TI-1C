## Mengatasi Missing Values dengan Drop Data

1. Hapus Data yang kosong (NaN)

```python
# Menghapus baris yang memiliki nilai kosong
df_drop = df.dropna()
df_drop
```

## Mengatasi Missing Values dengan Nilai Baru

2. Ganti Data yang kosong (NaN) menjadi :

*italicized text*<ul><li>Umur = 50 </li><li>Gaji = 25000000</li></ul>

```python
# Membuat salinan dataframe untuk mengisi nilai baru
df_fill_new = df.copy()

# Mengisi NaN pada 'Umur' dengan 30
df_fill_new['Umur'] = df_fill_new['Umur'].fillna(30)

# Mengisi NaN pada 'Gaji' dengan 25000000
df_fill_new['Gaji'] = df_fill_new['Gaji'].fillna(25000000)

df_fill_new
```

## Mengatasi Missing Values dengan Nilai Mean

3. Ganti Data Kosong (NaN) pada Umur dan Gaji dengan Nilai Rata-Rata

```python
df_fill_mean = df.copy()

# Mengisi NaN dengan nilai rata-rata (mean)
df_fill_mean['Umur'] = df_fill_mean['Umur'].fillna(df_fill_mean['Umur'].mean())
df_fill_mean['Gaji'] = df_fill_mean['Gaji'].fillna(df_fill_mean['Gaji'].mean())

df_fill_mean
```

```python
df_fill_mean = df.copy()

df_fill_mean['Umur'] = df_fill_mean['Umur'].fillna(df_fill_mean['Umur'].mean())
df_fill_mean['Gaji'] = df_fill_mean['Gaji'].fillna(df_fill_mean['Gaji'].mean())

# Bulatkan Umur ke integer, supaya sesuai contoh
df_fill_mean['Umur'] = df_fill_mean['Umur'].round(0).astype(int)
df_fill_mean['Gaji'] = df_fill_mean['Gaji'].round(0).astype(int)

df_fill_mean
```

## Mengatasi Missing Values dengan Nilai Median

4. Ganti Data Kosong (NaN) pada Umur dan Gaji dengan Nilai Median

```python
df_fill_median = df.copy()
df_fill_median['Umur'] = df_fill_median['Umur'].fillna(df_fill_median['Umur'].median())
df_fill_median['Gaji'] = df_fill_median['Gaji'].fillna(df_fill_median['Gaji'].median())
df_fill_median
```

## Mengatasi Missing Values dengan Nilai Mode

5. Ganti Data Kosong (NaN) pada Umur dan Gaji dengan Nilai Modus

```python
df_fill_mode = df.copy()

umur_mode = df_fill_mode['Umur'].mode()[0]
gaji_mode = df_fill_mode['Gaji'].mode()[0]

df_fill_mode['Umur'] = df_fill_mode['Umur'].fillna(umur_mode)
df_fill_mode['Gaji'] = df_fill_mode['Gaji'].fillna(gaji_mode)

df_fill_mode
```

## Mengatasi Wrong Data dengan Ubah Data

6. Apabila Ada Umur yang diatas 120 maka ubah menjadi 33

```python
# Membuat salinan dataframe untuk memperbaiki umur
df_fix_age = df.copy()

# Mengganti umur yang lebih dari 120 menjadi 33
df_fix_age.loc[df_fix_age['Umur'] > 120, 'Umur'] = 33

df_fix_age
```

## Soal-Soal

### Soal 1 : Buatlah table list Top 5 Customers yang memiliki nilai transaksi paling banyak

```python
# Soal 1 - Top 5 Customers berdasarkan total nilai transaksi

# Hitung total transaksi per customer
top_customers = final_df.groupby('CustomerID')['Price'].sum().sort_values(ascending=False).head(5)

# Ambil hanya 1 transaksi (pertama) dari masing-masing customer top
top_customers_info = final_df[final_df['CustomerID'].isin(top_customers.index)]

# Drop duplikat agar hanya 1 baris per customer
top_customers_info = top_customers_info.sort_values(by='CustomerID').drop_duplicates(subset='CustomerID', keep='first')

# Jadikan CustomerID sebagai index
top_customers_info.set_index('CustomerID', inplace=True)

# Tampilkan hasil akhir
top_customers_info
```

### Soal 2 : Buatlah list Top 5 Gas stations yang memiliki nilai transaksi paling banyak !

```python
# Soal 2
# Buatlah list Top 5 Gas stations yang memiliki nilai transaksi paling banyak

# Group by GasStationID, lalu hitung total Amount dan total Price
top_gasstations = final_df.groupby('GasStationID').agg({
    'Amount': 'sum',
    'Price': 'sum'
})

# Rename kolom agar sesuai contoh
top_gasstations.rename(columns={'Price': 'Gas Sales'}, inplace=True)

# Sort berdasarkan Gas Sales tertinggi
top_gasstations = top_gasstations.sort_values(by='Gas Sales', ascending=False).head(5)

# Tampilkan hasil
top_gasstations
```

### Soal 3 : Buatlah list Top 5 Jenis produk yang memiliki nilai transaksi paling banyak !

```python
# Soal 3
# Buatlah list Top 5 Jenis produk yang memiliki nilai transaksi paling banyak

# Group by Description (nama produk), lalu hitung total Amount dan total Price
top_products = final_df.groupby('Description').agg({
    'Amount': 'sum',
    'Price': 'sum'
})

# Rename kolom agar sesuai
top_products.rename(columns={'Price': 'Sales'}, inplace=True)

# Sort berdasarkan Sales tertinggi
top_products = top_products.sort_values(by='Sales', ascending=False).head(5)

# Tampilkan hasil
top_products
```

### Soal 4 : Buatlah deskripsi statistik untuk masing masing hari (23,24,25,26) pada dataset transaction_1k !

```python
# Soal 4 - Statistik gabungan Amount & Sales dalam satu tabel

# Pastikan Date bertipe datetime
final_df['Date'] = pd.to_datetime(final_df['Date'])

# Ambil hanya tanggal 23–26 Agustus
filtered = final_df[final_df['Date'].dt.day.isin([23, 24, 25, 26])]
filtered['Day'] = filtered['Date'].dt.date  # Buat kolom tanggal bersih

# Group by tanggal, lalu ambil describe untuk Amount & Price
desc = filtered.groupby('Day')[['Amount', 'Price']].describe()

# Rename Price jadi Sales (agar sesuai PDF)
desc.columns.names = ['Variable', 'Stat']
desc.rename_axis("Date", inplace=True)
desc.rename(columns={"Price": "Sales"}, level=0, inplace=True)

# Tampilkan hasil
desc
```

### Soal 5 : Carilah waktu terbaik (hari dan jam) dimana paling banyak user gas station melakukan transaksi !

```python
# Soal 5
# Carilah waktu terbaik (hari dan jam) dimana paling banyak user gas station melakukan transaksi

# Pastikan kolom Time bertipe datetime
final_df['Time'] = pd.to_datetime(final_df['Time'], format='%H:%M:%S')

# Buat kolom baru 'Hour' untuk jam transaksi
final_df['Hour'] = final_df['Time'].dt.hour

# Grouping berdasarkan Date dan Hour
day_hour_group = final_df.groupby([final_df['Date'].dt.date, 'Hour']).agg({
    'Amount': 'sum',
    'Price': 'sum'
})

# Rename kolom Price menjadi Gas Sales
day_hour_group.rename(columns={'Price': 'Gas Sales'}, inplace=True)

# Tampilkan tabel gabungan Amount dan Gas Sales per Date dan Hour
day_hour_group

# Untuk mencari waktu terbaik (transaksi terbanyak)
# Cari maksimum Amount per tanggal
best_times = day_hour_group.groupby(level=0).apply(lambda x: x.loc[x['Amount'].idxmax()])

# Tampilkan waktu terbaik
best_times
```
