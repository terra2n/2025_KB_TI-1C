# I. Latihan - Membersihkan Data Menggunakan Pendas

## Import Library

```python
import pandas as pd
```

## Load Dataset

Load Dataset

```python
df = pd.read_csv('./sample_data/Data (1).csv')
df
```

Mengecek Missing Values

```python
df.isnull().sum()
```

## Mengatasi Missing Values dengan Drop Data

1. Hapus Data yang kosong (NaN)

```python
# Menghapus baris yang memiliki nilai kosong
df_drop = df.dropna()
df_drop
```

## Mengatasi Missing Values dengan Nilai Baru

2. Ganti Data yang kosong (NaN) menjadi :

*italicized text*<ul><li>Umur = 50 </li><li>Gaji = 25000000</li></ul>

```python
# Membuat salinan dataframe untuk mengisi nilai baru
df_fill_new = df.copy()

# Mengisi NaN pada 'Umur' dengan 30
df_fill_new['Umur'] = df_fill_new['Umur'].fillna(30)

# Mengisi NaN pada 'Gaji' dengan 25000000
df_fill_new['Gaji'] = df_fill_new['Gaji'].fillna(25000000)

df_fill_new
```

## Mengatasi Missing Values dengan Nilai Mean

3. Ganti Data Kosong (NaN) pada Umur dan Gaji dengan Nilai Rata-Rata

```python
df_fill_mean = df.copy()

# Mengisi NaN dengan nilai rata-rata (mean)
df_fill_mean['Umur'] = df_fill_mean['Umur'].fillna(df_fill_mean['Umur'].mean())
df_fill_mean['Gaji'] = df_fill_mean['Gaji'].fillna(df_fill_mean['Gaji'].mean())

df_fill_mean
```

```python
df_fill_mean = df.copy()

df_fill_mean['Umur'] = df_fill_mean['Umur'].fillna(df_fill_mean['Umur'].mean())
df_fill_mean['Gaji'] = df_fill_mean['Gaji'].fillna(df_fill_mean['Gaji'].mean())

# Bulatkan Umur ke integer, supaya sesuai contoh
df_fill_mean['Umur'] = df_fill_mean['Umur'].round(0).astype(int)
df_fill_mean['Gaji'] = df_fill_mean['Gaji'].round(0).astype(int)

df_fill_mean
```

## Mengatasi Missing Values dengan Nilai Median

4. Ganti Data Kosong (NaN) pada Umur dan Gaji dengan Nilai Median

```python
df_fill_median = df.copy()
df_fill_median['Umur'] = df_fill_median['Umur'].fillna(df_fill_median['Umur'].median())
df_fill_median['Gaji'] = df_fill_median['Gaji'].fillna(df_fill_median['Gaji'].median())
df_fill_median
```

## Mengatasi Missing Values dengan Nilai Mode

5. Ganti Data Kosong (NaN) pada Umur dan Gaji dengan Nilai Modus

```python
df_fill_mode = df.copy()

umur_mode = df_fill_mode['Umur'].mode()[0]
gaji_mode = df_fill_mode['Gaji'].mode()[0]

df_fill_mode['Umur'] = df_fill_mode['Umur'].fillna(umur_mode)
df_fill_mode['Gaji'] = df_fill_mode['Gaji'].fillna(gaji_mode)

df_fill_mode
```

## Mengatasi Wrong Data dengan Ubah Data

6. Apabila Ada Umur yang diatas 120 maka ubah menjadi 33

```python
# Membuat salinan dataframe untuk memperbaiki umur
df_fix_age = df.copy()

# Mengganti umur yang lebih dari 120 menjadi 33
df_fix_age.loc[df_fix_age['Umur'] > 120, 'Umur'] = 33

df_fix_age
```
